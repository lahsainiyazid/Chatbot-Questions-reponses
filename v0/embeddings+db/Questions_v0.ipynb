```python
import json
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.documents import Document
from langchain_community.vectorstores import Chroma
with open("/content/chunks_questions.json","r",encoding="utf-8") as f:
  data=json.load(f)
docs=[]
for item in data:
  text=f"""
  Content:{item.get("page_content","")}
  """
  docs.append(Document(page_content=text,
                       metadata={"file_name":item.get("file_name",""),
                                 "source":item.get("source",""),
                                 "page":item.get("page",""),
                                 "folder_date":item.get("folder_date",""),
                                 "start_index":item.get("start_index","")}))
embeddings=HuggingFaceEmbeddings(model="BAAI/bge-m3",model_kwargs={"device":"cuda"},
                                 encode_kwargs={"normalize_embeddings":True})
batch_size=100
db=None
for i in range (0,len(docs),batch_size):
  print(f"Processing batch :{i}->{i+batch_size}")
  batch=docs[i:i+batch_size]
  db=Chroma.from_documents(documents=batch,embedding=embeddings,persist_directory="/content/db",
                           collection_name="db_questions_0")
```
